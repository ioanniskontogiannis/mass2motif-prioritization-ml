The goal of this notebook is to check which compounds are in MSnLib for positive ionization that were used to run MS2LDA and later on build the motifset that was used for machine learning.

The aim now is to remove all the spectra from GNPS that represent these compounds, in order to be sure that when we run MS2LDA with the GNPS dataset, curate the new motifs, we are not adding redundancy into the model.

```python
from matchms.importing import load_from_mgf
```

```python
msn_neg_path = '/home/ioannis/thesis_data/MSn_filtering_data/msn_negative_filtered.mgf'
gnps_neg_path = '/home/ioannis/thesis_data/gnps_negative_filtered.mgf'
```

```python
msn_spectra = list(load_from_mgf(msn_neg_path))
print('MSn pos Spectra: ', len(msn_spectra))

gnps_spectra = list(load_from_mgf(gnps_neg_path))
print('GNPS pos Spectra', len(gnps_spectra))
```

```python
unique_inchikeys_msn = {s.get("inchikey") for s in msn_spectra if s.get("inchikey")}
print("Unique InChIKeys from msn:", len(unique_inchikeys_msn))

unique_inchikeys_gnps = {s.get("inchikey") for s in gnps_spectra if s.get("inchikey")}
print("Unique InChIKeys from gnps:", len(unique_inchikeys_gnps))
```

```python
gnps_not_in_msn = unique_inchikeys_gnps - unique_inchikeys_msn
print("GNPS InChIKeys NOT in MSnLib:", len(gnps_not_in_msn))
```

```python
different_spectra = []
for s in gnps_spectra:
    if s.get('inchikey') in unique_inchikeys_msn:
        continue
    else:
        different_spectra.append(s)
print(len(different_spectra))
```

```python
from matchms.exporting import save_as_mgf
output_path = '/home/ioannis/thesis_data/GNPS_without_mnslib_neg.mgf'
save_as_mgf(different_spectra, output_path)
print(len(different_spectra))
```
